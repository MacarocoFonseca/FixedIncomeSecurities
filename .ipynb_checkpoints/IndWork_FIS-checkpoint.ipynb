```python
import numpy as np  
import pandas as pd
import QuantLib as ql
import matplotlib.pyplot as plt
import time 
```

```python
notional = 1000
couponrate = 0.0375
couponfreq = 2

issuedate = ql.Date(31,7,2015)
firstcoupon = ql.Date(15,1,2016)
maturity = ql.Date(15,7,2025)
settlementdate = ql.Date(25,10,2015)

infIssue = 237.14365
infSettlement = 237.46721
```

```python
u = ql.ActualActual().dayCount(issuedate, settlementdate)
w = ql.ActualActual().dayCount(issuedate, firstcoupon)
print(u,w)
```

```python
AI = (notional * (couponrate/couponfreq) * (u/w))
print(AI)

infVariation = (infSettlement / infIssue) - 1

newNotional = notional * (1+infVariation)

newAI = (newNotional * (couponrate/couponfreq) * (u/w))

print(newAI)
```

### b)

```python
#CPI Simulation 

import numpy.random as npr

npr.seed(1)
miu = 0.03513
sigma = 0.022439
sim = 1000
M = ql.ActualActual().dayCount(issuedate, maturity)
dt = 1./ M
S = np.zeros((M + 1, sim))
s0 = 237.14355
scpi_start_time = time.time()

S[0,:] = s0
S[1:,:] = S[0,:] * np.exp((miu - 0.5 * sigma ** 2 / 2) * dt + sigma * np.sqrt(dt) * np.cumsum(npr.randn(M,sim), 0))
    


dates = []

for idx in range(M + 1):
    dates.append(issuedate + idx)
sim_cpi = pd.DataFrame({'Date' : dates})
for idx in range(sim):
    sim_cpi.loc[:,'Simulation' +str(idx)] = S[:, idx]


sim_cpi
```

```python
print('CPI Simulation completed')
#print(sim, 'Simulations were executed in', "%s seconds" % (time.time() - sim_start_time), '.')
```

```python
ax = sim_cpi.plot(title = 'CPI Curve' , figsize = (20,7), legend=None)
plt.ylabel('CPI (%)')
plt.xlabel('Days')
```

```python
#Inflation Simulation

I = np.zeros((M, sim ))
sim_inflation = pd.DataFrame()
#sim_start_time = time.time()

for idx in range(M):
    sim_inflation.loc[idx, 'Date'] = sim_cpi.loc[idx, 'Date'] 
    for n in range(sim):
        sim_inflation.loc[idx, 'Simulation' + str(n)] = np.log(
                                                         (sim_cpi.loc[idx+1, 'Simulation' + str(n)]
                                                         / 
                                                          sim_cpi.loc[idx, 'Simulation' + str(n)]
                                                         ) \
                                                        ) \
                                                        * 100
print('Inflation Simulation completed')
#print(sim, 'Simulations were executed in', "%s seconds" % (time.time() - sim_start_time), '.')
```

```python
ax = sim_cpi.plot(title = 'Inflation Curve' , figsize = (12,4), legend=None)
plt.ylabel('Inflation (%)')
plt.xlabel('Days')





```
